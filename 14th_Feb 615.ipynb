```python
d={"A":[1,2,np.nan],"B":[5,np.nan,np.nan],"C":[1,2,3],"D":[np.nan,np.nan,np.nan]}
```

```python
df1=pd.DataFrame(d)
df1
```

```python
# To drop the values with Nan
#dropna()
df1.dropna()
```

```python
# dropna has 4 parameters (how,thresh,axis,inplace)
# how="all" or "any"
df1.dropna(how="all"  # all means it will be checking the rows with all nan value to delete that row)
```

```python
df1.dropna(how="any")        # any means if any one of the Nun is present in the rows it will be deleted
```

```python
df1.dropna(how="all",axis=1)
```

```python
df1.dropna(how="any",axis=1)
```

```python
# To check the sum of Nan values in each column
df1.isna()      #Nan=True
```

```python
df1.isna().sum()
```

```python
# thresh
df1.dropna(thresh=2)    # Min 2 values should be present
```

```python
df1.dropna(thresh=1)  #each row having min 1 number or more than one should be present
```

```python
df1.dropna(thresh=3)
```

```python
import numpy as np
x_years=np.arange(2000,2005,dtype=int)
y_innova=np.array([20,40,10,14,23])
y_honda_city=np.array([10,5,12,40,32])
x_years,y_innova
```

```python
plt.plot(x_years,y_innova)
```

```python
plt.figure(figsize=(4,4))
plt.plot(x_years,y_innova,'-+g')
plt.plot(x_years,y_honda_city,"--y")
plt.xlabel("Years")
plt.ylabel("No of cars")
plt.title("Car statistics")
```
